```python
fig = plt.figure(figsize=(10,10))

plt.scatter(df.CREDIT_LIMIT, df.PAYMENTS, edgecolor = 'black', s = 90)
plt.title('USERS LIMITS & PAYMENTS DETAILS')
plt.xlabel('CREDIT_LIMIT')
plt.ylabel('PAYMENTS')
circle1 = plt.Circle(centroid1, radius=0.33, edgecolor = 'black', fc='yellow')
circle2 = plt.Circle(centroid2, radius=0.95, edgecolor = 'black', fc='red')
circle3 = plt.Circle(centroid3, radius=0.95, edgecolor = 'black', fc='green')
plt.gca().add_patch(circle1)
plt.gca().add_patch(circle2)
plt.gca().add_patch(circle3)
```
